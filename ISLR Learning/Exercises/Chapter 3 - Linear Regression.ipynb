#### The diagnostic plots include 4 plots

```python
import statsmodels.api as sm
from statsmodels.graphics.gofplots import ProbPlot

```

```python
#First Plot: Residual vs Fitted: examine nonlinear relationship

y_pred = lm1_fit.predict(sm.add_constant(auto['horsepower']))

_ = sns.residplot(y_pred,'mpg',data = auto,lowess = True,
                 line_kws = {'color':'red','ls':'--','alpha':.5},
                 scatter_kws = {'alpha':.8})

plt.xlabel('Fitted Values')
plt.ylabel('Residual')
plt.title('Residuals vs Fitted')
```

- From the plots above, we can see that there is a clear non-linearity to the residual line versus the fitted values.

```python
from scipy.stats import zscore
# Normla Q-Q Plot: Check whehter residuals are normally distributed

lm1_resid = lm1_fit.resid #model residuals

lm1_norm_resid = zscore(lm1_resid) #Normalized residuals

lm1_resid_stud = lm1_fit.get_influence().resid_studentized_internal # studentized residuals

leverage = lm1_fit.get_influence().hat_matrix_diag
```

```python
QQ = ProbPlot(lm1_norm_resid)

plot = QQ.qqplot(line = '45',alpha = .5,color = 'blue',lw = 1)
plt.title('Normal Q-Q Plot')

# annotations
abs_norm_resid = np.flip(np.argsort(np.abs(lm1_norm_resid)),0)
top3 = abs_norm_resid[:3]
least2 = abs_norm_resid[-2:]
for r,i in enumerate(top3):
    plt.annotate(i,xy = (np.flip(QQ.theoretical_quantiles,0)[r],lm1_norm_resid[i]))
```

- Looking at the graph above, there are several points that fall far away from the red line. This is indicative of the errors not being normally distributed. That is to say, We are more likely to see extreme values than to be expected if the data was truly normally distributed.

```python
#3. Scale location: this plot is to check homoscedasticity & outlier
sns.residplot(y_pred,lm1_resid_stud,data = auto,
              lowess = True,
              line_kws ={'color':'red','alpha':.8,'ls':'--'},scatter_kws = {'alpha': .5} )
plt.axhline(y = 3,color = 'black',ls = '--',alpha = .3)
plt.axhline(y = -3,color = 'black',ls = '--',alpha = .3)
plt.title('Check for Homoscedasticity')
plt.xlabel('Fitted Value')
plt.ylabel('Studentized Residuals & Outliers')
```

```python
#4. Check for leverage vs residuals
model_cooks = lm1_fit.get_influence().cooks_distance[0] #points >.5 have high leverage
_ = sns.residplot(leverage,lm1_resid_stud,lowess = True,
                 line_kws = {'color':'red','alpha':.5,'ls':'--'})
top3_leverage = np.flip(np.argsort(model_cooks),0)[:3]
for i in top3_leverage:
    plt.annotate(i,xy = (leverage[i],
                        lm1_resid_stud[i]))
    
    
#Based on the book, the average value of leverage statistic (hi) is (p+1)/n. The leverage value larger than this might be
#suspected has high leverage
high_lev = (1+1)/372
plt.axvline(x = high_lev,c = 'blue',ls ='--',alpha = .6)

plt.title('Residual vs Leverage')
plt.xlabel('Leverage')
plt.ylabel('Studentized Residual')
```

### (d) Use the plot() function to produce diagnostic plots of the linear regression fit. Comment on any problems you see with the fit. Do the residual plots suggest any unusually large outliers? Does the leverage plot identify any observations with unusually high leverage?

```python
def plot(fitted_model,y,data):
    y_pred = fitted_model.predict()#Calculate predict values
    norm_resid = fitted_model.get_influence().resid_studentized_internal #normalized residuals
    norm_abs_resid = np.sqrt(np.abs(norm_resid)) #Plot square root of the absolute normalized residuals
    leverage = fitted_model.get_influence().hat_matrix_diag
    
    fig,ax = plt.subplots(2,2,tight_layout = True,figsize = (10,6))
    
    axes = ax.flatten()
    #Plot Residual against Fitted Values
    ax1 = sns.residplot(y_pred,y,data = data,
                     lowess = True,
                     line_kws = {'color':'red','ls':'--','alpha':.5},
                     scatter_kws = {'alpha':.5},ax = axes[0])
    ax1.set_title('Residual vs Fitted')
    ax1.set_xlabel('Fitted')
    ax1.set_ylabel('Residual')
    
    #Plot Residual QQ Plot
    QQ = ProbPlot(norm_resid)
    ax2 = QQ.qqplot(QQ,line = '45',alpha = .5,color = 'blue',lw = 1,ax = axes[1])
    
    #Plot Scale location
    axes[2].scatter(y_pred,norm_abs_resid,color = 'blue',alpha = .5)
    ax3 = sns.regplot(y_pred,norm_abs_resid,lowess = True,
                       line_kws = {'color':'red','ls':'--','alpha':.5},ax = axes[2],
                       scatter = False)
    ax3.set_title('Scale Location: Check Heterodesdacity')
    ax3.set_ylabel('Absolute Standardized Residuals')
    ax3.set_xlabel('Fitted Value')
    
    #Plot Leverage
    axes[3].scatter(leverage,norm_resid,alpha = .5)
    ax4 = sns.regplot(leverage,norm_resid,
                        lowess = True,
                       line_kws = {'color':'red','ls':'--','alpha':.5},
                     scatter_kws = {'alpha':.5},ax = axes[3])
    ax4.set_xlim([0,.3])
    ax4.set_title('Leverage Statistic vs Standardized Residual')
    ax4.set_xlabel('Leverage')
    ax4.set_ylabel('Standardized Residuals')
```

```python
columns = [i for i in list(boston.columns) if i not in(['crim'])]


lm_boston = [smf.ols(formula = f'crim ~ {c}',data = boston).fit() for c in columns]

dp = 4
for l in lm_boston:
    display(pd.DataFrame({'coef': np.around(l.params,dp),
                         'std_err': np.around(l.bse,dp),
                         't': np.around(l.tvalues,dp),
                         'P>|t|': np.around(l.pvalues,dp),
                         '[.025': np.around(l.conf_int(alpha =.05)[0],dp),
                         '.975]': np.around(l.conf_int(alpha = .05)[1],dp)}))
    
           
```

```python
sig_p = [l.params[1:].index[0] for l in lm_boston if l.pvalues[1]<=.05]
nonsig_p = [l.params[1:].index[0] for l in lm_boston if l.pvalues[1]>.05]

print(f'Sig Predictors include: {sig_p}')

print(f'Non-sig P include: {nonsig_p}')
```
