In addition, the variance of X for these two sets of companies was ˆσ2 = 36. Finally, 80% of companies issued dividends. Assuming that X follows a normal distribution, predict the probability that a company will issue a dividend this year given that its percentage profit was X = 4 last year.

|

- Based on Bayes' theorem, I have the following equation:

- P (Y = 1|X = 4) = p(X = 4|Y = 1)p(y=1)/p(X = 4) , where the numerators are all known and the denominator is: p(x = 4|y = 0)p(y = 0) + p(x = 4|y = 1)p(y = 1)

```python
import scipy.stats as ss

#1. P(y = 1) = .8 and p(y = 0) = .2
py = .8
qy = .2

#2. P(X =4|Y = 1): probability of an observation in a normal distribution with mean =10 and varaince = 36
p_x4_1 = ss.norm(10,6).pdf(4)
p_x4_0 = ss.norm(0,6).pdf(4)

P = p_x4_1*py/(p_x4_0*qy + p_x4_1*py)

print('The probability is: ',P)
```

- Build a Logistic Regression Models

```python
X = boston.drop('resp',axis = 1)
y = boston['resp']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 1,test_size = .3)
#Fit the raw predictors
b_lr = smf.logit('resp ~ dis+ptratio+tax+age+nox+indus',data = boston.iloc[X_train.index]).fit()
b_lr.summary()
```

- 3 out of 6 predictors had non-significant 
- Based on the distribution, all predictors are not normally distributed, and the scale are quite different. Therefore, a scale transformation will be used to see the performance of the lr

```python
#Check accuracy
def check_accuracy(ytrue,ypred):
    print('By fitting a LR, the accuracy of the classifier is:')
    print(classification_report(ytrue,ypred))
    print('\n')
    print('The accuracy score is:',np.round(accuracy_score(ytrue,ypred),2))
```

```python
def roc_curve_plot(ytrue,ypred,label):
    fpr,tpr,threshold =roc_curve(ytrue,ypred)
    _ = plt.plot(fpr,tpr,label = label)
    _ = plt.plot([0,1],ls = '--',alpha = .5,color = 'gray')
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC Curve')
    plt.legend(loc = 'best')
```

```python
roc_curve_plot(y_test,b_lr.predict(X_test),'LR')
```

```python
b_lr_pred = np.where(b_lr.predict(X_test)<.5,0,1)
check_accuracy(y_test,b_lr_pred)
confusion_matrix(y_test,b_lr_pred)
print('F1 Score is:',f1_score(y_test,b_lr_pred))
```
