#### Helper Function for Linear model fit (Sklearn)

```python

def linear_fit(x,y,f,err_var,numP):
    linear_fit = smf.ols(formula = f,data = pd.concat([x,y],axis = 1)).fit()
    RSS = linear_fit.ssr
    Cp = 1/len(y)*(RSS+2*numP*err_var)
    adj_R2 = linear_fit.rsquared_adj
    AIC = linear_fit.aic
    BIC = linear_fit.bic
    return Cp, adj_R2,AIC,BIC
```

```python
#Create the dataset
df = pd.DataFrame({'X':X,
                   'X2':X**2,
                  'X3':X**3,
                  'X4':X**4,
                  'X5':X**5,
                  'X6':X**6,
                  'X7':X**7,
                  'X8':X**8,
                  'X9':X**9,
                  'X10':X**10,
                  'y':Y})

df.head()
```

### (d) Repeat (c), using forward stepwise selection and also using backwards stepwise selection. How does your answer compare to the results in (c)?

```python
def linear_fit(x,y,f,err_var,numP):
    lfit = smf.ols(formula = f,data = pd.concat([x,y],axis = 1)).fit()
    RSS = lfit.ssr
    Cp = 1/len(y)*(RSS+2*numP*err_var)
    BIC = lfit.bic
    AIC = lfit.aic
    adj_R2 = lfit.rsquared_adj
    return Cp,BIC,AIC,adj_R2,RSS
```

```python
x_f = df.drop('y',axis = 1)
y_f = df['y']
var = list(x_f.columns) #List all possible variables
feature_list,Cp_list,BIC_list,AIC_list,adj_R2list,RSS_list = [],[],[],[],[],[]
features = []
#Fit the model
for i in range(x_f.shape[1]):
    best_RSS = np.inf
    for comb in itertools.combinations(var,1):
        xf = ' + '.join(feature_list+[comb[0]])
        f = f'y ~ {xf}'
        tmp_result = linear_fit(x_f,y_f,f,err_var,len(feature_list+[comb[0]]))
        RSS_tmp = tmp_result[4]
        if RSS_tmp < best_RSS: #Select
            best_RSS = RSS_tmp
            best_feature = comb[0]
            best_Cp= tmp_result[0]
            best_BIC = tmp_result[1]
            best_AIC = tmp_result[2]
            best_adjR2 = tmp_result[3]
            
         #Updating the next loop
    var.remove(best_feature) 
    feature_list.append(best_feature)
    features += [feature_list.copy()]
    Cp_list.append(best_Cp)
    BIC_list.append(best_BIC)
    AIC_list.append(best_AIC)
    adj_R2list.append(best_adjR2)
    RSS_list.append(best_RSS)

df_forward = pd.DataFrame({'Features':features,
                          'Cp_score': Cp_list,
                          'BIC': BIC_list,
                          'AIC': AIC_list,
                          'adj_R2': adj_R2list,
                          'RSS': RSS_list})    
```

### Fit a backward stepwise selection

```python
def linear_fitb(data,f,err_var,numP):
    lb_fit = smf.ols(formula = f,data = data).fit()
    RSS = lb_fit.ssr
    Cp_score = 1/len(df)*(RSS+2*numP-1*err_var)
    BIC = lb_fit.bic
    AIC = lb_fit.aic
    adj_R2 = lb_fit.rsquared_adj
    return RSS,BIC,AIC,adj_R2,Cp_score
```

```python
#Conduct backward stepwise selection
x_b = df.drop('y',axis = 1)
y_b = df['y']
var = list(x_b.columns)
full = ' + '.join(var)
full_fit = linear_fitb(df,f"y ~ {full}",err_var,len(var))

feature_list,Cp_list,BIC_list,AIC_list,adj_R2_list,RSS_list = [list(x_b.columns)],[full_fit[4]],[full_fit[1]],[full_fit[2]],[full_fit[3]],[full_fit[0]]

for i in range(len(var)-1):
    best_RSS = np.inf
    
    for comb in itertools.combinations(var,1):
        features = var.copy()
        features.remove(comb[0])
        xmb = ' + '.join(features)
        tmp_b = linear_fitb(df,f'y ~ {xmb}',err_var,len(features))
        RSS = tmp_b[0]
        if RSS < best_RSS:
            best_RSS = RSS
            worst_feature = comb[0]
            best_BIC = tmp_b[1]
            best_AIC = tmp_b[2]
            best_adj_R2 = tmp_b[3]
            best_Cp_score = tmp_b[4]
            
    #Remove the selected variable and cont
    var.remove(worst_feature)
    feature_list+=[var.copy()]
    Cp_list.append(best_Cp_score)
    BIC_list.append(best_BIC)
    AIC_list.append(best_AIC)
    adj_R2_list.append(best_adj_R2)
    RSS_list.append(best_RSS)
    
df_b = pd.DataFrame({'Features':feature_list,
                    'Cp_score':Cp_list,
                    'BIC':BIC_list,
                    'AIC':AIC_list,
                    'adj_R2':adj_R2_list,
                    'RSS':RSS_list})
        
```

```python
#Conduct bset subset selection
df2 = pd.DataFrame({'x1':x,
                  'x2':x**2,
                  'x3':x**3,
                  'x4':x**4,
                  'x5':x**5,
                  'x6':x**6,
                  'x7':x**7,
                  'x8':x**8,
                  'x9':x**9,
                  'x10':x**10,
                  'y':y})

display(df2.head())
```

```python
df_x = df2.drop('y',axis = 1)
df_y = df2['y']
var = list(df_x.columns)            
```

```python
#Fit a full model
full_var = ' + '.join(var)
full_fit = smf.ols(formula = f'y ~ {full_var}',data = df2).fit()
full_RSS = full_fit.ssr
err_var = 1/(len(df2)-10-1)*full_RSS
err_var
```

```python
def linear_fit(data,f,err_var,numP):
    lr_fit = smf.ols(formula = f,data = data).fit()
    RSS = lr_fit.ssr
    BIC = lr_fit.bic
    AIC = lr_fit.aic
    adj_R2 = lr_fit.rsquared_adj
    Cp_score = 1/len(data)*(RSS + 2*numP*err_var)
    return RSS,BIC,AIC,adj_R2,Cp_score
```

```python
RSS_list,BIC_list,AIC_list,adj_R2_list,Cp_score_list = [],[],[],[],[]
feature_list = []
num_features = []

for k in tnrange(1,10+1,desc = 'Loop...'):
    for comb in itertools.combinations(var,k):
        x_comb = ' + '.join(comb)
        f = f'y~{x_comb}'
        fit_test = linear_fit(df2,f,err_var,len(comb)) #Fit model with the different subsets
        RSS_list.append(fit_test[0])
        BIC_list.append(fit_test[1])
        AIC_list.append(fit_test[2])
        adj_R2_list.append(fit_test[3])
        Cp_score_list.append(fit_test[4])
        num_features.append(len(comb))
        feature_list.append(comb)
            
            
df7 = pd.DataFrame({'RSS':RSS_list,
                   'BIC':BIC_list,
                   'AIC':AIC_list,
                   'adj_R2':adj_R2_list,
                   'Cp_score':Cp_score_list,
                   'Num_Features': num_features,
                   'Features':feature_list})
            
            
```

### (e) Fit a PCR model on the training set, with M chosen by crossvalidation. Report the test error obtained, along with the value of M selected by cross-validation.

```python
#Generate an instance of PCA
pca = PCA()

X_pca = pca.fit_transform(scale(X_train)) #Obtain the principle components for scaled X variables
n = len(X_pca)

#10-fold CV, with shuffle
kf_10 = KFold(n_splits = 10, shuffle = True,random_state = 1)
mse = []

#Calculate MSE with only intercept (no principle components in regression)
regr = LinearRegression()
score = -1*cross_val_score(regr,np.ones((n,1)),y_train,cv = kf_10,scoring = 'neg_mean_squared_error').mean()
mse.append(score)

#Calculate MSE using CV, adding one component at a time
for i in range(1,len(list(X_train.columns))+1):
    score = -1*cross_val_score(regr,X_pca[:,:i],y_train,cv = kf_10,scoring = 'neg_mean_squared_error').mean()
    mse.append(score)
    
plt.plot(np.array(mse),'-v')
plt.xlabel('Number of Principle Components in Regression')
plt.ylabel('MSE')
plt.title('Scree Plot')
plt.xlim((0,18))
```

- With the Scree Plot, it can be seen that M = 2 is at the elbow, so the first 2 components will be selected

```python
#Apply PCA on test dataset
X_pca_test = pca.transform(scale(X_test))[:,:2]


pcr = LinearRegression()
pcr_2 = pcr.fit(X_pca[:,:2],y_train)


pcr2_predict = pcr_2.predict(X_pca_test)


mse2 = mean_squared_error(y_test,pcr2_predict)

print('The RMSE by fitting a PCR with 2 components is: ', np.sqrt(mse2))
```

- It seems that using only 2 components yielded larger test error than using all components (i.e., Linear Regression)
